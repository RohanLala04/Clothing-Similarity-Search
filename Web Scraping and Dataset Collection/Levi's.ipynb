### Levi Women

```python
import selenium
import time
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import time

# Function to crawl a web page and return the parsed HTML using Selenium and BeautifulSoup
def crawler(link, user_agent, chromedriver_path):
    # Configure Selenium options
    options = webdriver.ChromeOptions()
    options.headless = True
    options.add_argument(f'user-agent={user_agent}')
    options.add_argument("--window-size=1920,1080")
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--allow-running-insecure-content')
    options.add_argument("--disable-extensions")
    options.add_argument("--proxy-server='direct://'")
    options.add_argument("--proxy-bypass-list=*")
    options.add_argument("--start-maximized")
    options.add_argument('--disable-gpu')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--no-sandbox')
    
    # Create a WebDriver instance with the configured options
    driver = webdriver.Chrome(executable_path=chromedriver_path, options=options)
    
    # Wait for a short duration to ensure the WebDriver is ready
    time.sleep(0.5)
    
    # Open the specified link in the WebDriver
    driver.get(link)
    
    # Scroll down the page in increments of 100 pixels
    height = driver.execute_script("return document.body.scrollHeight")
    for scrol in range(100, height, 100):
        driver.execute_script(f"window.scrollTo(0, {scrol})")
        time.sleep(0.3)
    
    # Get the page source after scrolling
    src = driver.page_source
    
    # Parse the HTML source using BeautifulSoup
    tango = bs(src.encode("utf-8"), "html")
    
    # Quit the WebDriver instance
    driver.quit()
    
    # Return the parsed HTML
    return tango

# Input parameters
link = 'https://www.levi.in/sale/women-2?pmin=1%2C000.00&pmax=3%2C000.00&prefn1=items_type&prefv1=Jeans&sz=224'
user_agent = "YOUR_USER_AGENT"
chromedriver_path = r'C:\Program Files (x86)\chromedriver'

# Call the crawler function
soup = crawler(link, user_agent, chromedriver_path)

# Find the elements containing product titles and links
product_titles = soup.find_all('a', class_='name-link')
product_links = ["https://www.levi.in" + link['href'] for link in product_titles] 
# Extract the text from the product titles
product_titles_text = [title.text.strip() for title in product_titles]

# Print or further process the extracted data
for title, link in zip(product_titles_text, product_links):
    print(f"Product Title: {title}")
    print(f"Product Link: {link}")
    print()
```

```python
import pandas as pd

# Create a DataFrame from the lists
df = pd.DataFrame({'Product_Description': product_titles_text, 'Product_Link': product_links})

df.tail()
```

```python
# Save the DataFrame to a CSV file
df.to_csv('levi-W_output.csv', index=False)
```
